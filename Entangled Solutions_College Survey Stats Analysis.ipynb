```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_excel('LUMINA_non_response.xlsx', sheet_name = 0, header =1, skiprows=0)

data.drop(['Screen', 'Job 1','Job 2','Job 3','Job 4','Job 5', 'Control','Unnamed: 9','Unnamed: 15', 'Unnamed: 21',
           'Unnamed: 27','Unnamed: 33', 'Unnamed: 55','Unnamed: 56','Unnamed: 57','Unnamed: 58','Unnamed: 59','Unnamed: 60',
           'Unnamed: 61','Unnamed: 62','Unnamed: 63','Unnamed: 64', 'Unnamed: 65','Unnamed: 66','Unnamed: 67',
           'Unnamed: 68','Unnamed: 69','Unnamed: 70','Unnamed: 71'
          ],axis=1,inplace =True)
data = data.convert_objects(convert_numeric=True)

data = data.rename(columns={'Unnamed: 49': 'PPAGE', 'Unnamed: 50': 'ppagecat','Unnamed: 51':'ppagect4',
                            'Unnamed: 52':'PPEDUC', 'Unnamed: 53': 'PPEDUCAT','Unnamed: 54': 'PPETHM' })

data.head()



```

# Correlation Rates

```python
price_block = data['Price.1']
alpha_quality_A = data['Quality A.1']


def correlation(x,y):
    std_x = (x-x.mean())/x.std(ddof=0)
    std_y = (y-y.mean())/y.std(ddof=0)
    
    return (std_x * std_y).mean()

print (correlation(price_block, alpha_quality_A))
```


# Aggregated Dataset

```python
data1 = data[['Price','Quality A', 'Value A', 'Quality B','Value B']]

data2 = data[['Price.1','Quality A.1','Value A.1','Quality B.1','Value B.1']]
data3 = data[['Price.2', 'Quality A.2','Value A.2','Quality B.2', 'Value B.2']]
data4 = data[['Price.3', 'Quality A.3', 'Value A.3','Quality B.3', 'Value B.3']]
data5 = data[['Price.4', 'Quality A.4', 'Value A.4','Quality B.4', 'Value B.4']]


data2= data2.rename(columns={'Price.1': 'Price', 'Quality A.1': 'Quality A','Value A.1':'Value A',
                            'Quality B.1':'Quality B', 'Value B.1': 'Value B' })
data3= data3.rename(columns={'Price.2': 'Price', 'Quality A.2': 'Quality A','Value A.2':'Value A',
                            'Quality B.2':'Quality B', 'Value B.2': 'Value B' })
data4= data4.rename(columns={'Price.3': 'Price', 'Quality A.3': 'Quality A','Value A.3':'Value A',
                            'Quality B.3':'Quality B', 'Value B.3': 'Value B' })
data5= data5.rename(columns={'Price.4': 'Price', 'Quality A.4': 'Quality A','Value A.4':'Value A',
                            'Quality B.4':'Quality B', 'Value B.4': 'Value B' })

agg_data = data1.append([data2,data3,data4,data5])

agg_data.head(5)
```
